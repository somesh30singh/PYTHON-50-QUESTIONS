```python
def find_decimal( decimal : int) -> str:
    """
    Converts a given decimal number to its binary representation using recursion.

    Parameters:
            n (int): A non-negative integer to be converted into binary.

    Returns:
        str: Binary representation of the given decimal number.

    """
    if not isinstance(decimal, int):
        raise TypeError("Input must be an integer")
    if decimal == 0:
        return 0
    result = ""
    while decimal > 0:
        digit = decimal % 2
        decimal = decimal // 2
        result = str(digit) + result
    return result

if __name__ == "__main__":
    try:
        user_input = input("ENTER YOUR NUMBER :")
        if not user_input.isdigit():
            raise TypeError("ERROR : PLEASE ENTER VALID INPUT")
        user_input = int(user_input)

        result = find_decimal(user_input)
        print(result)

    except TypeError as error:
        print(error)
    except ValueError as error:
        print(error)
    except Exception as error:
        print(error)
```
