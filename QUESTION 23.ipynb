# Question 23 : Find Intersection of Two Lists

```python
list1 = ["a","b","c","d"]
list2 = ["c","d","e","f"]
result = []
for item in list1:
    if item in list2 and item not in result:
        result.append(item)
print(result)
```

```python
def intersection_number(list1 : list, list2: list) -> list:
    """
        Find the common elements (intersection) between two lists.
    
        The function returns a list containing elements that appear in both input lists.
        Duplicate handling depends on implementation (can be unique or repeated).
    
        Parameters:
        list1 (list): First list.
        list2 (list): Second list.
    
        Returns:
        list: A list of common elements.
    
        Example:
        [1, 2, 3] & [2, 3, 4] -> [2, 3]
    """

    if not isinstance(list1,list) or not isinstance(list2,list):
        raise TypeError("Input must be a list")
            
    if len(list1) == 0 or len(list2) == 0:
        raise ValueError("Input should not be empty")

    result = []
    for item in list1:
        if item in list2 and item not in result:
            result.append(item)
    return result

if __name__ =="__main__":
    try:
        list1_input = input("PLEASE ENTER YOUR LIST :")
        list2_input = input("PLEASE ENTER YOUR LIST :")
        
        list1_input = [int(item) for item in list1_input.split()]
        list2_input = [int(item) for item in list2_input.split()]
    except ValueError:
        print("ERROR : PLEASE ENTER VALID INPUT")
    else:
        try:
            result = intersection_number(list1_input,list2_input)
            print(result)
            
        except (TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
