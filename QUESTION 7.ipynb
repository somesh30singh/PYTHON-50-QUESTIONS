# Question 7: Count Vowels in a String

```python
string = "this is python class"
vowels = "aeiou"
count = 0
for char in string.lower():
    if char in vowels:
        count = count + 1
print(count)
```

```python
def count_vowels(text : str) -> int:
    """find the vowels from the given userinput

    return :
           it will count and return vowels
    raises:
          TypeError: input should be a string
          ValueError : input should not be empty
    """

    if not isinstance(text,str):
        raise TypeError("input must be a string")
    if text.strip() == "":
        raise ValueError("input should not be empty")
    else:
        text = text.strip().lower()

    Vowels = "aeiou"
    count = 0
    for chara in text.lower():
        if chara in Vowels:
            count = count + 1
    return count

if __name__ == "__main__":
    try:
        user_input = input("ENTER SENTENCE : ")
        result = count_vowels(user_input)
        print(f" Total Vowels are {result}")
    except TypeError as error:
        print(error)
    except ValueError as error:
        print(error)
    except Exception as error:
        print(error)

```
