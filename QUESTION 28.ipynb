```python
def list_merge(list1 : list, list2: list) -> list:
    """
        Merge two sorted lists into a single sorted list.
    
        This function takes two lists that are already sorted in ascending order
        and merges them into one sorted list, also in ascending order.
    
        Parameters:
        list1 (list): First sorted list
        list2 (list): Second sorted list
    
        Returns:
        list: A new sorted list containing all elements from both input lists
    
        Example:
        >>> merge_sorted_lists([1, 3, 5], [2, 4, 6])
        [1, 2, 3, 4, 5, 6]
    """
    if not isinstance(list1,list) or not isinstance(list2,list):
        raise TypeError("INPUT MUST BE A LIST ONLY")
    for item in list1 or list2:
        if not isinstance(item,(int)):
            raise TypeError("INPUT OF BOTH LIST SHOULD BE INTEGER VALUE ONLY")
    if len(list1) == 0 and len(list2) == 0:
        raise ValueError("LIST SHOULD NOT BE EMPTY")

    list_merging = []
    i=j=0
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            list_merging.append(list1[i])
            i = i + 1
        else:
            list_merging.append(list2[j])
            j = j + 1
            
    while i < len(list1):
        list_merging.append(list1[i])
        i = i + 1
    while j < len(list2):
        list_merging.append(list2[j])
        j = j + 1
    return list_merging

if __name__ == "__main__":
    try:
        list_input1 = input("ENTER LIST :")
        list_input2 = input("ENTER LIST :")

        list_input1 = [int(item) for item in list_input1.split()]
        list_input2 = [int(item) for item in list_input2.split()]

    except ValueError:
        print("ERROR : PLEASE ENTER VALID INPUT")
    else:
        try:
            result = list_merge(list_input1,list_input2)
            print("LIST MERGED SUCCESSFULLY :",result)
        except(TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
