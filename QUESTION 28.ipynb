# Question 28: Merge Two Sorted Lists

```python
list1 = [1,2,3,4,5,6]
list2 = [5,6,7,8,9]

i=j=0

merge_list = []
while i <len(list1) and j < len(list2):
    if list1[i] <= list2[j]:
        merge_list.append(list1[i])
        i = i + 1
    else:
        merge_list.append(list2[j])
        j = j +1
        
while i<len(list1):
    merge_list.append(list1[i])
    i = i +1

while j<len(list2):
    merge_list.append(list2[j])
    j = j +1

print(merge_list)
```
