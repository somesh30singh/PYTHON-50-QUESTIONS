# Question 17: Generate List of Even Numbers

```python
number =  20
even_number = []
for i in range(2,number+1,2):
    even_number.append(i)
print(even_number)
```

```python
def Even_number_list(numb : int) -> int:
    """Generate a list of even numbers up to the given limit.

    This function returns all even numbers from 0 to n (inclusive).

    Args:
        n (int): The upper limit.

    Returns:
        list: A list containing even numbers up to n.

    Raises:
        TypeError: If n is not an integer.
        ValueError: If n is negative.
    """

    if not isinstance(numb,int):
        raise TypeError("Input mjust be an integer")
    if numb <= 2:
        raise ValueError("Input should be greater then 2")

    Even_Numb = []
    for num in range(2,numb+2,2):
        Even_Numb.append(num)
    return Even_Numb
        

if __name__ == "__main__":
    try:
        user_input = input("PLEASE ENTER YOUR NUMBER")
        user_input = int(user_input)

    except ValueError:
        print("Please Enter Valid Input")

    else:
        try:
            result =  Even_number_list(user_input)
            print("LIST OF EVEN NUMBERS :",result)
        except(TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
