# Question 4 : Find all prime numbers between 100 to 200

```python
number = 23
for num in range(2,int(number**.5)+1):
    if number % num == 0:
        print(f"{number} is not prime number")
        break
else:
    print(f"{number} is prime number")
```

```python
def find_prime_numbers(numbers : int) -> bool:
    """ This function will find the prime number from the given list
    Args: 
        (numbers : int) -> bool:, input must be an integer and work on boolean
    return:
          it will return all the prime numbers.
    raise:
         TypeError: input must be an integer
         ValueError: input should not be empty
    
    """
    if not isinstance(numbers,int):
        raise TypeError("input must be an integer")
    if numbers < 2:
        raise ValueError("input should be greater or equal to 2")

    for numb in range(2,int(numbers**.5)+1):
        if numbers % numb == 0:
            return False
    else:
        return True

if __name__ == "__main__":
    try:
        prime_number_list = []
        starting_limit = input("ENTER NUMBER :")
        ending_limit = input("ENTER NUMBER :")
        if not starting_limit.strip().isdigit:
            raise TypeError("input must be a Numeric")
        if not ending_limit.strip().isdigit:
            raise TypeError("input must be a Numeric")

        starting_limit = int(starting_limit)
        ending_limit = int(ending_limit)

        for num1 in range(starting_limit,ending_limit):
            if find_prime_numbers(num1):
                print(num1,"prime number")
                prime_number_list.append(num1)
        print(prime_number_list)

    except (TypeError,ValueError) as error:
        print("input should be numeric only")
    except Exception as error:
        print("Something went wrong")
```
