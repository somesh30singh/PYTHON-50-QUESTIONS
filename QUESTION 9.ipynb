# Question 9: find sum of all digit using loop

```python
number = 123
total = 0
while number > 0:
    digit = number % 10
    total = total + digit
    number = number // 10
print(total)
```

```python
def sum_of_digit(numbers : int) -> int:
    """calculate the total number and find sum of all digit

    args:
        int: input must be an integers

    return:
         This will return sum of all digit given by users

    raises:
          TypeError: Input must be an integer.
          ValueError: input should not be negative or single digit.
    """

    if not isinstance(numbers, int):
        raise TypeError("input should be integers")
    if numbers < 0:
        raise ValueError("Input should not be negative")
    for i in range(10):
        if numbers <= i:
            raise ValueError("input should be atleast 2 digit")

    total = 0
    while numbers > 0:
        digits = numbers % 10
        total =  total + digits
        numbers = numbers // 10
    return total

if __name__ == "__main__":
    user_input = input("ENTER YOUR NUMBER :")
    try:
        user_input = int(user_input)
    except ValueError:
        print("ERROR :","Please enter Valid input")
    else:
        try:
            result = sum_of_digit(user_input)
            print(f"sum of total digits are : {result}")
        except (TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
