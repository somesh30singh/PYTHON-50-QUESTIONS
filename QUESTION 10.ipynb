# Question 10: find duplicate value of given input 

```python
number = [1,12,2,2,2,28,4,5,5,5,56,6,8,7]
result = []
for num in number:
    if num not in result:
        result.append(num)
print(result)
```

```python
from typing import List,Any
def remove_duplicates(numbers : List[Any]) -> List[Any]:
    """Find duplicate values from the given input list.

    This function takes a list as input and returns a new list
    containing the values that appear more than once in the input.
    Each duplicate value is returned only once.

    Parameters:
        data (list): A list of elements in which duplicates need to be found.

    Returns:
        list: A list containing duplicate values.

    Raises:
        TypeError: If the input is not a list.
        ValueError : if input is empty or single element
    """

    if not isinstance(numbers, list):
        raise TypeError("Input must be a list")
    if len(numbers) == 0:
        raise ValueError("input should not be empty")
    if len(numbers) == 1:
        print("List has only single element")
        return numbers
        

    result = []
    for numb in numbers:
        if numb not in result:
            result.append(numb)
    return result

if __name__ == "__main__":
    Number_list = [1,1,1,2,2,3,3,5,5,6,6,4,4,5,5]
    try:
        result = remove_duplicates(Number_list)
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print(error)
    else:
        print("List after removing duplicates :",result)
```
