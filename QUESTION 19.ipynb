```python
def reverse_sentence(sentence : str) -> str:
    """
        Reverse the order of words in a sentence while keeping each word unchanged.

        The function splits the sentence into words, reverses their order,
        and joins them back into a single string.
    
        Parameters:
        sentence (str): The input sentence.
    
        Returns:
        str: A new sentence with the word order reversed.
    
        Example:
        "Hello world from Python" -> "Python from world Hello"
    """

    if not isinstance(sentence,str):
        raise TypeError("input must be a string")
    if len(sentence) == 0:
        raise ValueError("Input should not be empty")

    word_list = []
    word = ""

    for char in sentence:
        if char != " ":
            word = word + char
        else:
            word_list.append(word)
            word = ""
    word_list.append(word)

    rev_sen = ""
    for item in range(len(word_list)-1,-1,-1):
        rev_sen = rev_sen + word_list[item]

        if item != 0:
            rev_sen = rev_sen + " "
    return rev_sen

if __name__ == "__main__":
    try:
        user_input = input("ENTER YOUR SENTENCE :")

        if user_input.isdigit():
            raise TypeError("Input should be a string only")

        result = reverse_sentence(user_input)
        print(result)
    except TypeError as error:
        print(error)
    except ValueError as error:
        print(error)
    except Exception as error:
        print(error)
```
