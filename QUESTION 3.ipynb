# Question 3: Find the prime number

```python
number = 23
for i in range(2,int(number**.5)+1):
    if number % i == 0:
        print(f"{number} is not prime number")
        break
else:
    print(f"{number} is prime number")
```

```python
def find_prime_number(number1 : int) -> bool:
    """ This function will find wether the number is prime or not
    args:
        int: (number1 : int) -> bool:
    Return : it will return anser in boolean value

    raise:
         TypeError: input must be an integer
         ValueError: input should not be empty
    """

    if not isinstance(number1,int):
        raise TypeError("input must be integer")
    if number1 < 2:
        raise ValueError("input should not be empty")

    for num in range(2,int(number1**.5)+1):
        if number1 % num == 0:
            return False
            break
    else:
        return True

if __name__ == "__main__":
    try:
        input_user = input("ENTER NUMBER")
        if not input_user.strip().isdigit():
            raise TypeError("input must be a digit")
        number1 = int(input_user)
        if find_prime_number(number1):
            print(f"{number1} is prime number")
        else:
            print(f"{number1} is not prime number")
    except (TypeError,ValueError) as error:
        print(error)
    except Exception  as error:
        print("Something went wrong")
```
