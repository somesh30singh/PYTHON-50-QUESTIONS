# Question 6: Find the given user input is palindrom or not.

```python
# Palindrom

string = "mom"
rev = ""
for chara in string:
    rev = chara + rev
    
print(rev)
```

```python
def is_palindrom_or_not(text : str) -> bool:
    """Find is string value is palindrom or not
     return:
           it will revser the string and find whether the user_input is palindrom or not
    raise:
        TypeError: input must be a string
        ValueError:input should not be empty
    """

    if not isinstance(text, str):
        raise TypeError("input must be a string")
    if text.strip() == "":
        raise ValueError("input should not be empty")
    else:
        text = text.strip().lower()

    reverse = ""
    for char in text:
        reverse = char + reverse
    return text == reverse

if __name__ == "__main__":
    try:
        user_input = input("Enter alphabet").strip().lower()
        if is_palindrom_or_not(user_input):
            print(f"'{user_input}' is a palindrom")
        else:
            print(f"'{user_input}' is not a palindrom")
    except TypeError as error:
        print("Input must be an alphabet")
    except ValueError as error:
        print("Input should not be empty")
    except Exception as error:
        print(error)
```

```python
is_palindrom_or_not("mom")
```
