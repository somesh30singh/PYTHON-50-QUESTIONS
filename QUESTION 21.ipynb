```python
def GCD(a : int, b : int) -> int:
    """
        Find the Greatest Common Divisor (GCD) of two numbers using recursion.
    
        This function uses the Euclidean algorithm, where gcd(a, b) = gcd(b, a % b).
        The recursion continues until b becomes 0.
    
        Parameters:
        a (int): First number.
        b (int): Second number.
    
        Returns:
        int: GCD of the two numbers.
    
        Example:
        gcd(12, 18) -> 6

        TypeError:
        raise TypeError("Input must be an integer in both object")

    """
    if not isinstance(a, int) or not isinstance(b,int):
        raise TypeError("Input must be an integer in both object")
    if a == 0 and b == 0:
        raise ValueError("Input should not be zero")
     
    if b == 0:
        return a
    return GCD(b, a%b)

if __name__ == "__main__":
    try:
        user_input1 = input("ENTER YOUR NUMBER :")
        user_input2 = input("ENTER YOUR NUMBER :")

        user_input1 = int(user_input1)
        user_input2 = int(user_input2)

    except ValueError:
        print("ERROR : PLEASE ENTER VALID INPUT")

    else:
        try:
            result = GCD(user_input1,user_input2)
            print(result)

        except TypeError as error:
            print(error)
        except ValueError as error:
            print(error)
        except Exception as error:
            print(error)
```
