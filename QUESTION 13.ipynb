# Question 13 find given input is anagram or not?

```python
string1 = "abc"
string2 = "abc"
char_count = {}

if len(string1) != len(string2):
    print(False)
else:
    for char in string1.lower().replace(" ",""):
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    for char in string2.lower().replace(" ",""):
        if char not in char_count:
            print(False)
            break
        char_count[char] -= 1
        if char_count[char] < 0:
            print(False)
            break
    else:
        print(True)
```

```python
def Is_anagram_or_not(string3 :str, string4: str) -> dict:
    """Check whether the given two input strings are anagrams or not.

    Two strings are considered anagrams if they contain the same
    characters with the same frequency, ignoring case and spaces.

    Args:
        text1 (str): First input string.
        text2 (str): Second input string.

    Returns:
        bool: True if both strings are anagrams, otherwise False.

    Raises:
        TypeError: If any input is not a string.
        ValueError: If any input string is empty.
    """

    if not isinstance(string3,str) or not isinstance(string4,str):
        raise TypeError("input must be a string")

    string3 = string3.lower().replace(" ","")
    string4 = string4.lower().replace(" ","")

    if len(string3) != len(string4):
        return False

    text_count = {}
    for text in string3:
        if text in text_count:
            text_count[text] += 1
        else:
            text_count[text] = 1

    for text in string4:
        if text not in text_count:
            return False    
        text_count[text] -= 1
        if text_count[text] > 0:
            return False
    else:
        return True

if __name__ == "__main__":
    try:
        user_input = input("Please enter :")
        user_input = input("Please enter :")
        result = Is_anagram_or_not(user_input,user_input)
        if result:
            print(f"{result}, This is anagram")
        else:
            print(f"{result}, This is not anagram")
    except TypeError as error:
        print(error)
    except ValueError as error:
        print(error)
```
