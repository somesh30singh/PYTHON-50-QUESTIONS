```python
def calculate_GCD(numb1 : int, numb2 : int) -> int:
    """
        Compute the Greatest Common Divisor (GCD) of two integers.

        The function returns the largest number that divides both inputs
        without leaving a remainder.
    
        Parameters:
        a (int): First number.
        b (int): Second number.
    
        Returns:
        int: GCD of the two numbers.
    
        Example:
        gcd(12, 18) -> 6

        TypeError:
        raise TypeError("Input must be an integer in both object")
    """
    if not isinstance(numb1, int) or not isinstance(numb2,int):
        raise TypeError("Input must be an integer in both object")
    if numb1 == 0 and numb2 == 0:
        return 0


    while numb2 !=0:
        numb1,numb2 =  numb2,numb1% numb2
    return numb1

if __name__ == "__main__":
    try:
        user_input1 = input("ENTER YOUR NUMBER :")
        user_input2 = input("ENTER YOUR NUMBER :")

        if not user_input1.isdigit():
            raise TypeError("ERROR : PLEASE ENTER VALID INPUT")
        if not user_input2.isdigit():
            raise TypeError("ERROR : PLEASE ENTER VALID INPUT")
        user_input1 = int(user_input1)
        user_input2 = int(user_input2)
    
        result = calculate_GCD(user_input1,user_input2)
        print(result)

    except TypeError as error:
        print(error)
    except ValueError as error:
        print(error)
    except Exception as error:
        print(error)
```
