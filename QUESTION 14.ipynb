# Question 14: Calculate Fibonacci Sequence (Iterative)

```python

n = 10
a = 0
b = 1
for i in range(n):
    print(a,end = " ")
    a,b = b, a+b
```

```python
def fibonacci_sequence(number: int) -> None:
    """Calculate Fibonacci sequence using an iterative approach.

    The Fibonacci sequence is a series of numbers where each number
    is the sum of the two preceding numbers, starting from 0 and 1.

    Args:
        n (int): Number of terms to generate.

    Returns:
        list: A list containing the Fibonacci sequence up to n terms.

    Raises:
        TypeError: If n is not an integer.
        ValueError: If n is less than or equal to 0.
    """

    if not isinstance(number, int):
        raise TypeError("Input must be an integer")
    if number <= 0:
        raise ValueError("integer should not be negative or zero")

    num = 0
    numb = 1
    for i in range(number):
        print(num, end =" ")
        num, numb = numb, num+numb

if __name__ == "__main__":
    try:
        user_input = input("ENTER YOUR NUMBER :")
        user_input = int(user_input)
    except ValueError:
        print("Please Enter Valid input")
    else:
        try:
            fibonacci_sequence(user_input)
        except(TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
