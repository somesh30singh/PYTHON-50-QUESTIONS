```python
 def Find_Leap_year(Year: int) -> bool:
    """Check whether the given input year is a leap year or not.

    A year is considered a leap year if:
    - It is divisible by 4, AND
    - Not divisible by 100, UNLESS it is also divisible by 400.

    Args:
        year (int): The year to check.

    Returns:
        bool: True if the year is a leap year, otherwise False.

    Raises:
        TypeError: If the input is not an integer.
        ValueError: If the year is less than or equal to 0.
    """

    if not isinstance(Year,int):
        raise TypeError("Input must be an integer")
    if Year < 1582:
        raise ValueError("Year should be greater than 1582")

    if Year >= 1582 and(Year % 4 == 0 and Year % 100 != 0) or (Year % 400 == 0):
        return True
    else:
        return False

if __name__ == "__main__":
    try:
        user_input = input("Please entter your year :")
        user_input = int(user_input)
    except ValueError:
        print("Please Enter Valid Input")
    else:
        try:
            result = Find_Leap_year(user_input)
            if result:
                print(f"{result} : Yes, This is Leap year")
            else:
                print((f"{result} : No, This is not Leap year"))
        except(TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
