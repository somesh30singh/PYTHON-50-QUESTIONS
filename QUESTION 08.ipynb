```python
from typing import List
def find_second_largest_number(numbers : list[int]) -> int:
    """The Function will return second largest number from list
    Args:
        Input should be integer
    return:
         it will return the second largest number from list

    raises:
          TypeError: input must be a list of integers
          ValueError :input should contains atleast 2 elemets in list
    
    """
    if not isinstance(numbers, list):
        raise TypeError("input must be a list of integers")
    if any (not isinstance(num,int) for num in numbers):
        raise TypeError("All elements must be integer if list")
    if len(numbers) < 0:
        raise ValueError("list mist contain atleast integer 2 elements")

    if numbers[0] > numbers[1]:
        largest = numbers[0]
        second = numbers[1]
    else:
        largest = numbers[1]
        second = numbers[0]
    for numb in numbers[2:]:
        if numb > largest:
            second = largest
            largest = numb
        elif numb != largest and numb > second:
            second = numb
    if largest == second:
        raise ValueError("list must contain atleast 2 unique elements")
    return second

if __name__ == "__main__":
    user_input = [1,2,3,5,4,8,7,11,9,12]
    try:
        result = find_second_largest_number(user_input)
    except TypeError as error:
        print(error)
    except ValueError as error:
        print(error)
    except Exception as error:
        print(error)
    else:
        print("second largest number :", result)
```
