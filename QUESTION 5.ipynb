# Question 5: Find the factorial number of user input

```python
# Find Factorial
number = 5
fact = 1
for num in range(1,number +1):
    fact = fact * num
print(fact)
```

```python
def factorial_finder(numbers :int) -> int:
    """This function will find the factorial number
    args:
        int : (numbers :int) -> int:, input should be an integer
    return:
          it will return fazctorial number

    raise:
         TypeError: input must be an integer
         ValueError : input should not be empty
    
    """
    if isinstance(numbers, bool):
        raise TypeError("Boolean Value is not acceptable")
    if not isinstance(numbers , int):
        raise TypeError("Input should be a integer")
    if numbers < 0:
        raise ValueError("Input should not be negative")

    factorial = 1
    for numb in range(1,numbers+1):
        factorial = factorial * numb
    return factorial

if __name__ == "__main__":
    try:
        User_input = input("ENTER NUMBER : ")
        User_input = int(User_input)
        print("Factorial Number :",factorial_finder(User_input))
    except TypeError as error:
        print("Input must be a numeric")
    except ValueError as error:
        print("Input should not be negative")
    except Exception as error:
        print("Something went wrong")
```
