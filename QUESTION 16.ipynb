# Calculate power without using ** operator

```python
base_value =  5
power_value = 5
result = 1
for i in range(power_value):
    result = result * base_value
print(result)
```

```python
def Calculate_power(Base_value : int, Power_value: int ) -> int:
    """Calculate the power of a number without using the ** operator.

    This function computes base raised to the given exponent using
    iterative multiplication.

    Args:
        base (float | int): The base number.
        exponent (int): The exponent value.

    Returns:
        float: Result of base raised to the exponent.

    Raises:
        TypeError: If base is not numeric or exponent is not an integer.
        ValueError: If exponent is negative (when negative exponents are not supported).
    """

    if not isinstance(Base_value,int) or not isinstance(Power_value,int):
        raise TypeError("Both power and base value should be integer")
    if Power_value < 0:
        raise ValueError("Power value should not be negative")

    result = 1
    for i in range(Power_value):
        result = result * Base_value
    return result

if __name__ == "__main__":
    try:
        base_input = input("ENTER BASE NUMBER :")
        power_input = input("ENTER POWER NUMBER")
        base_input = int(base_input)
        power_input =  int(power_input)

    except ValueError:
        print("ERROR:", "Please enter valid input")

    else:
        try:
            result = Calculate_power(base_input,power_input)
            print(f" Power of {base_input} is : {result} ")

        except(TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
