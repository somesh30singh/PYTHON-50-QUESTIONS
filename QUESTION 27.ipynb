# Question 27: Check if List is Sorted

```python
sorted_list1 = [ 1,2,3,4,5,6]

for i in range(0,len(sorted_list1)-1):
    if sorted_list1[i] > sorted_list1[i+1]:
        print(False)
        break
else:
    print(True)
```

```python
def sorted_or_not(sorted_list : list) -> bool:
    """
        Check if a list is sorted in ascending order.

        The function verifies whether each element in the list is less than
        or equal to the next element.
    
        Parameters:
        lst (list): The input list.
        
        Returns:
        bool: True if the list is sorted in ascending order, otherwise False.
    
        Example:
        [1, 2, 3] -> True
        [3, 2, 1] -> False
    
    """
    if not isinstance(sorted_list,list):
        raise TypeError("Input must be a list")
    if len(sorted_list) == 0:
        raise ValueError("Input should not be empty")

    for i in range(len(sorted_list)-1):
        if sorted_list[i] > sorted_list[i+1]:
            return False
            break
            
    return True

if __name__ == "__main__":
    try:
        user_input1 = input("PLEASE ENTER LIST :")        

        user_input1 = [int(item) for item in user_input1.split()]
        result = sorted_or_not(user_input1)
        print(result)

    except(TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print(error)
```
