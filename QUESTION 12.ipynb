# Question 12: Count characters of a given string

```python
string1 = "my name is somesh and im learning python sql x"
char_count = {}
for char in string1.lower().replace(" ",""):
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```

```python
def characters_count(text : str) -> str:
    """Count the total number of characters in a given string.

    Args:
        text (str): The input string whose characters need to be counted.

    Returns:
        int: Total number of characters in the string.

    Raises:
        TypeError: If the input is not a string.
        ValueError : If the input is empty.
    """
    if not isinstance(text, str):
        raise TypeError("input must be a string")
    if len(text) == 0:
        raise ValueError("input should not be empty")

    text_count = {}
    for chara in text.lower().replace(" ",""):
        if chara in text_count:
            text_count[chara] += 1
        else:
            text_count[chara] = 1
    return text_count

if __name__ == "__main__":
    try:
        user_input = input("Enter your sentence :")
        result = characters_count(user_input)
        print("CHARACTERS COUNT LIST :", result)
    except (TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print(error)
```
