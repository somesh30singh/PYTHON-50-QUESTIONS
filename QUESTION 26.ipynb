```python
def sorted_list(my_list : list) -> list:
    """
        Calculate the sum of a list of numbers using recursion.
    
        The function adds the first element of the list to the sum
        of the remaining elements recursively until the list is empty.
    
        Parameters:
        numbers (list): List of numeric values.
    
        Returns:
        int/float: Sum of all elements in the list.
    
        Example:
        [1, 2, 3, 4] -> 10
    """
    if not isinstance(my_list, list):
        raise TypeError("Input should be a list")
    if len(my_list) == 0:
        raise ValueError("Input should not be empty")
        
    length = len(my_list)
    for i in range(length):
        for j in range(0,length -i -1):
            if my_list[j] > my_list[j+1]:
                my_list[j],my_list[j+1] = my_list[j+1],my_list[j]
    return my_list

if __name__ == "__main__":
    try:
        user_list1 = input("ENTER :")
        user_list1 = [int(item) for item in user_list1.split()]

    except ValueError:
        print("ERROR : PLEASE ENTER VALID INPUT")

    else:
        try:
            result = sorted_list(user_list1)
            print("SORTED LIST:",user_list1)
        except(TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
