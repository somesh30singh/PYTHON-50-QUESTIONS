# Question 24 : Remove All Whitespace from a String

```python
string =  "               python            class                  test         question   24   "
result = ""
pre_space = False
for char in string.strip():
    if not char.isspace():
        result = result + char.capitalize()
        pre_space = False
    else:
        if pre_space == False:
            result = result + " "
            pre_space = True
print(result)
```

```python
def remove_whitespace(string1 : str) -> str:
    """ 
        Remove all whitespace characters (spaces, tabs, and newlines) from a string.

        The function processes the input string and returns a new string
        with all types of whitespace removed.
    
        Parameters:
        text (str): The input string.
    
        Returns:
        str: String without any whitespace characters.
    
        Example:
        "Hello World\n\t" -> "HelloWorld"
    """
    if not isinstance(string1, str):
        raise TypeError("Input must be a string")
    if len(string1) == 0:
        raise ValueError("Input should not be empty")

    result = ""
    pre_space = False
    for char in string1.strip():
        if not char.isspace():
            result = result + char.capitalize()
            pre_space = False
        else:
            if pre_space == False:
                result = result + " "
                pre_space = True
    return result

if __name__ == "__main__":
    try:
        user_input = input("ENTER :")
        if user_input.isdigit():
            raise TypeError("ERROR : PLEASE ENTER VALID INPUT")

        result = remove_whitespace(user_input)
        print(result)
        
    except(TypeError,ValueError) as error:
        print(error)
    except Exception as error:
        print(error)
```
