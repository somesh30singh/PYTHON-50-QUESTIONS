```python
def binary_implementation(sorted_list : list, target : int) -> int:
    """
    Performs binary search on a sorted list to find the target element.

    Parameters:
    arr (list): A sorted list of elements in ascending order.
    target (int/float): The element to search for in the list.

    Returns:
    int: The index of the target element if found.
         Returns -1 if the target is not present in the list.
    """
    if not isinstance(sorted_list, list):
        raise TypeError("Input must be a sorted list")
    if not isinstance(target,int):
        raise TypeError("Input must be an integer")
    if len(sorted_list) == 0:
        raise ValueError("Input should not be empty")


    low = 0
    high = len(sorted_list)-1
    while low <=high:
        mid = (low + high) //2
        if sorted_list[mid] == target:
            return mid
            
        elif target < sorted_list[mid]:
            high = mid -1
        else:
            low = mid +1
    else:
        return -1

if __name__=="__main__":
    try:
        sorted_list = input("ENTER SORTED LIST :")
        if not sorted_list:
            raise ValueError("List Should not be empty")
        sorted_list = [int(item) for item in sorted_list.split()]
            
        target1 = input("ENTER NUMBER TO FIND FROM THE LIST : ")
        if not target1:
            raise ValueError("Target Value should not be empty")
        target1 = int(target1)

    except ValueError:
        print("ERROR : PLEASE ENTER VALID INPUT")
    else:
        try:
            result = binary_implementation(sorted_list,target1)
            print(f"INDEX VALUE OF GIVEN TARGET : {result}")
        except (TypeError,ValueError) as error:
            print(error)
        except Exception as error:
            print(error)
```
